```python
#import the required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read the data from the csv
data = pd.read_csv("diabetes.csv")
#preview the first few rows of data
data.head()
```

```python
#explore the data based on the age and outcome
plt.scatter(data.Age, data.Outcome, marker='+', color='red')
```

```python
#import sklearn packages
from sklearn.model_selection import train_test_split
```

```python
#split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data[['Age']], data.Outcome, train_size=0.9)
```

```python
#preview the feature test data
X_test
```

```python
#preview the feature training data
X_train
```

```python
#preview the label training data
y_train
```

```python
#preview the label test data
y_test
```

```python
#import the sklearn linear model package
from sklearn.linear_model import LogisticRegression
```

```python
#create the model
model = LogisticRegression()
```

```python
#train the model
model.fit(X_train, y_train)
```

```python
#test the model using the test data
model.predict(X_test)
```

```python
#check the level of accuracy of model
model.score(X_test, y_test)
```

```python
#check probability of test data
model.predict_proba(X_test)
```

```python
#predict the outcome from new feature
age = np.array(5).reshape(-1, 1)
model.predict(age)
```
